```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

1. Створіть вектор розміром 11 зі значеннями від 0 до 5. Розрахуйте на його основі вектор y (y = x^2)

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
x
```

```python
y
```

2. Побудова графіків

2.1. Функціональний метод

```python
#Functional method
plt.plot(x, y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("First")
```

```python
#Create multiply plots
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

2.2. Об’єктно-орієнтований метод

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title("Second")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.1, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'r')
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title("First")

axes2.plot(x, y, 'b')
axes2.set_xlabel('X Label')
axes2.set_ylabel('Y Label')
axes2.set_title("Second")
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x, y, 'r')
axes[1].plot(y, x, 'b')
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label = "x**2")
ax.plot(x, x**3, label = "x**3")
ax.legend()
```

При створенні графіку можна задати наступні характеристики:
-	колір графіку 


```python
fig, axes = plt.subplots()
axes.plot(x, x**2, color = "y")
axes.plot(x, x**3, color = "g")
```

-	тип та ширину лінії

```python
fig, axes = plt.subplots()
axes.plot(x, x**2, lw = 2, ls = "--")
```

-	тип, розмір та колір маркерів

```python
fig, axes = plt.subplots()
axes.plot(x, x**2, marker = "o", markersize = 10, markerfacecolor = "r")
```
