```python
import seaborn as sns
%matplotlib inline
```

```python
planet = sns.load_dataset('planets')
planet.head()
```

```python
sns.distplot(planet['year'])
```

```python
sns.jointplot(x = "orbital_period", y = "mass", data = planet)
```

```python
sns.pairplot(planet)
```

```python
sns.pairplot(planet, hue = 'method')
```

```python
sns.barplot(x = 'orbital_period', y = 'year', data = planet)
```

```python
import numpy as np
```

```python
sns.barplot(x = 'orbital_period', y = 'year', data = planet, estimator = np.median)
```

```python
sns.boxplot(x = 'method', y = 'distance', data = planet, palette = 'rainbow')
```

```python
sns.stripplot(x = 'number', y = 'distance', data = planet)
```

```python
sns.heatmap(planet.corr())
```

```python
import matplotlib.pyplot as plt
grid = sns.FacetGrid(planet, col = 'method', row = 'number')
grid = grid.map(plt.hist, 'distance')
```
